## Задание 2

```python
t = np.linspace (0, 10 , 51)
f = np.cos(t)
fig, ax = plt.subplots()
ax.plot(t, f, color='Green')
plt.title('График f(t)')
ax.set_xlabel('Значения t')
ax.set_ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
```

## *Задание 3

```python
x = np.linspace(-3, 3, 51)
y1 = x**2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0, 0].plot(x, y1)
ax[0,0].set(title='График y1')

ax[0, 1].plot(x, y2)
ax[0,1].set(title='График y2')

ax[1, 0].plot(x, y3)
ax[1,0].set(title='График y3')

ax[1, 1].plot(x, y4)
ax[1,1].set(title='График y4')

ax[0, 0].axis(xmin= -5, xmax= 5)

fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
```
