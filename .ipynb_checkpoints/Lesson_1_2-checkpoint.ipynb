```python
import numpy as np
import pandas as pd
import pickle
```

```python
authors = pd.DataFrame({'author_id': [1, 2, 3], 'author_name': ['Тургенев', 'Чехов', 'Островский']})
```

```python
authors
```

```python
books = pd.DataFrame({'author_id': [1, 1, 1, 2, 2, 3, 3], 'book_title': ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'], 'price': [450, 300, 350, 500, 450, 370, 290]}, columns= ['author_id', 'book_title', 'price'])
```

```python
books
```

```python
authors_price = pd.merge(authors, books, on = 'author_id')
authors_price
```

```python
top5 = authors_price.copy()
top5 = top5.sort_values('price', ascending=False)
top5 = top5.head(5)
top5
```

```python
gvar_mi = authors_price.groupby('author_id').min()
gvar_ma = authors_price.groupby('author_id').max()
gvar_me = authors_price.groupby('author_id').mean()

authors_stat = pd.DataFrame({'author_name': gvar_mi['author_name'], 'min_price': gvar_mi['price'], 'max_price': gvar_ma['price'], 'mean_price': gvar_me['price']})

authors_stat
```

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
authors_price
```

```python
book_info = authors_price.pivot_table('price', index='author_name', columns='cover', aggfunc = 'sum', fill_value= 0)
book_info
```

```python
with open('book_info.pkl', 'wb') as file:
    pickle.dump(book_info, file)
```

```python
with open('book_info.pkl', 'rb') as file:
    book_info2 = pickle.load(file)
```
